```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Part 1.1 
Defining a function to read in csv files on German power companys wind turbine generation data. A column for the company name was added, and the date/time column was converted to date/time format.

```python
def intro(filename):
    df = pd.read_csv(filename)
    df.Date = pd.to_datetime(df.Date)
    df.insert(loc = 0, column = 'Company', value = (filename[:-4]))
    return(df)
              
```

## Part 1.2
Executing function from part 1.1 for each company's data

```python
hertz = intro('50Hertz.csv')
```

```python
amp = intro('Amprion.csv')
```

```python
tennet = intro('TenneTTSO.csv')
```

```python
trans = intro('TransnetBW.csv')
```

## Part 1.3
- creating a dataframe of all files concatenated (altog -> all together)

```python
altog = pd.concat([hertz,amp,tennet,trans])
```

Checking number of columns (should be 98)

```python
len(altog.columns)
```

## Part 1.4
Creating a dataset of the mean power generation for each company at each time interval and year.

```python
aggregate = altog.groupby([altog.Company, altog.Date.dt.to_period("Y")]).mean()
```

```python
aggregate
```

## Part 1.5
Creating a barh plat for the midnight and noon power generation in 2019 and 2020 for each company.

```python
midnight = aggregate['00:00:00']
noon = aggregate['12:00:00']
```

```python
midnight_barh = plt.figure()
midnight.sort_values().plot(kind='barh')

plt.title('TW/hour by Company/Year @ midnight')
```

```python
noon_barh = plt.figure()
noon.sort_values().plot(kind='barh', color='red')

plt.title('TW/hour by Company/Year @ noon')
```

## Part 1.6
The two plots were put together as subplot so they could be printed on one page.

```python
combined_barh, (ax_mid, ax_noon) = plt.subplots(2, 1, sharex=True)
midnight.sort_values().plot(kind='barh', ax=ax_mid, color='silver', title='TW/hour by Company/Year @ midnight')
noon.sort_values().plot(kind='barh', ax=ax_noon, color='gray', title='TW/hour by Company/Year @ noon')
plt.savefig('combined_bar_chart.png')
```

## Part 1.7
Creating a function that will accept the ungrouped data, a company name, a date, and a color: function will produce an area chart of the power generation for that day.

```python
def daily_view(dataframe, company, date, color='purple'):
    df = dataframe.query(f'Company=="{company}" and Date=="{date}"').drop(columns = ['Company','Date']).melt()

    fig = df.plot(
        kind = 'area',
        figsize=(8,4),
        x='variable',
        y='value',
        color=f'{color}',
        title = f'Power Production: {company}/{date}',
        legend=False
    )
    fig.set_xlabel('Time of Day')
    fig.set_ylabel('TW/h Generated')
```

```python
daily_view(altog, '50Hertz', '2019-08-24', 'purple')
```

## Part 1.8 (BONUS)
Creating a figure with a subplot for each company, showing power generation for the same day.

```python
a = hertz.query('Date=="2019-08-24"').drop(columns = ['Company','Date']).melt().rename(columns = {'variable' : 'Time of Day', 'value' : 'TW/h Generated'})
b = amp.query('Date=="2019-08-24"').drop(columns = ['Company','Date']).melt().rename(columns = {'variable' : 'Time of Day', 'value' : 'TW/h Generated'})
c = tennet.query('Date=="2019-08-24"').drop(columns = ['Company','Date']).melt().rename(columns = {'variable' : 'Time of Day', 'value' : 'TW/h Generated'})
d = trans.query('Date=="2019-08-24"').drop(columns = ['Company','Date']).melt().rename(columns = {'variable' : 'Time of Day', 'value' : 'TW/h Generated'})
```

```python
combo_area, ax = plt.subplots(2,2)

a.plot(kind='area', 
       ax=ax[0,0],
       figsize=(4,3), 
       color='red', 
       title='Power Production: 50Hertz/2019-08-24',
       x='Time of Day',
       y='TW/h Generated',
       legend=False)
ax[0,0].set_ylabel('TW/h Generated')
ax[0,0].set_xlabel('Time of Day')

b.plot(kind='area',
       ax=ax[0,1],
       figsize=(4,3),
       color='green',
       title='Power Production: Amprion/2019-08-24',
       x='Time of Day',
       y='TW/h Generated',
       legend=False)
ax[0,1].set_ylabel('TW/h Generated')
ax[0,1].set_xlabel('Time of Day')

c.plot(kind='area',
       ax=ax[1,0],
       figsize=(4,3),
       color='blue',
       title='Power Production: TenneTTSO/2019-08-24',
       x='Time of Day',
       y='TW/h Generated',
       legend=False)
ax[1,0].set_ylabel('TW/h Generated')
ax[1,0].set_xlabel('Time of Day')

d.plot(kind='area',
       ax=ax[1,1],
       figsize=(9,5),
       color='purple',
       title='Power Production: TransnetBW/2019-08-24}',
       x='Time of Day',
       y='TW/h Generated',
       legend=False)
ax[1,1].set_ylabel('TW/h Generated')
ax[1,1].set_xlabel('Time of Day')


combo_area.tight_layout(pad=3.0)
```
